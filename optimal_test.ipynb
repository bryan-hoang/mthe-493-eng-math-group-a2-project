```python
from src import optimal
import matplotlib.pyplot as plt
```

```python
results = optimal.test_random(10000, verbose=-1, log=False, n_factor=0.5)
```

```python
feasible = [x for x in results if x.feasible]
data = [(x.input_set, x.timing) for x in feasible]
data.sort(key=lambda d: d[1]["total"], reverse=True)
```

```python
n = [d[0].n for d in data]
k = [d[0].k for d in data]
time_keys = ["init", "get_employable_workers", "get_capacity", "assign", "reassign", "total"]
```

```python
fig, ax = plt.subplots(len(time_keys), 2, figsize=(10, 20))
for (i, key) in enumerate(time_keys):
    time_data = [d[1][key] for d in data]
    ax[i][0].scatter(k, time_data);  # k
    ax[i][0].set_xlabel("k")
    ax[i][0].set_ylabel("{} (ms)".format(key))
    ax[i][1].scatter(n, time_data);  # n
    ax[i][1].set_xlabel("n")
    ax[i][1].set_ylabel("{} (ms)".format(key))
fig.tight_layout(pad=3.0)
```
